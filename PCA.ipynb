```python
from sklearn import preprocessing
#each feature is scaled to [0,1] range
min_max_scaler = preprocessing.MinMaxScaler()
data_norm = min_max_scaler.fit_transform(data)
data_norm
```

b.Next, Perform Kmeans clustering (for this problem, use the Kmeans implementation in scikit-learn) on the
image data (since there are a total 7 pre-assigned image classes, you should use K = 7 in your clustering).
Use Euclidean distance as your distance measure for the clustering. Print the cluster centroids (use some
formatting so that they are visually understandable). Compare your 7 clusters to the 7 pre-assigned classes
by computing the Completeness and Homogeneity values of the generated clusters

```python
from sklearn.cluster import KMeans
k = 7
kmeans = KMeans(n_clusters = k, max_iter=500, verbose=1) #initialization
kmeans.fit(data_norm)
```

```python
names = pd.read_csv('segmentation_names.txt',header=None)
print(len(names))
```

```python
predicts = kmeans.labels_
```

```python
centroids = kmeans.cluster_centers_
```

```python
print "Cluster Centroids"
print pd.DataFrame(centroids, columns = ([i for i in names[0]]),
                   index = ('Cluster 1','cluster 2','cluster 3','cluster 4','cluster 5','cluster 6','cluster 7')).T               
```

```python
#compute completeness and Homogeneity
from sklearn.metrics import completeness_score,homogeneity_score
completeness_score(classes[1],predicts)
```

```python
homogeneity_score(classes[1],predicts)
```

c.Perform PCA on the normalized image data matrix. You may use the linear algebra package in Numpy or
the Decomposition module in scikit-learn (the latter is much more efficient). Analyze the principal
components to determine the number, r, of PCs needed to capture at least 95% of variance in the data.
Then use these rcomponents as features to transform the data into a reduced dimension space. 

```python
from sklearn import decomposition
pca = decomposition.PCA(n_components=7)
data_trans = pca.fit_transform(data_norm)
r = pca.explained_variance_ratio_
```

```python
r
```

```python
sum(r)
```

d.Perform Kmeans again, but this time on the lower dimensional transformed data. Then, compute the
Completeness and Homogeneity values of the new clusters.

```python
kmeans.fit(data_trans)
```

```python
new_centroids = kmeans.cluster_centers_
new_predicts = kmeans.labels_
```

```python
print "New Cluster Centroids"
print pd.DataFrame(new_centroids, columns = ('PC1','PC2','PC3','PC4','PC5','PC6','PC7'),
                   index = ('Cluster 1','cluster 2','cluster 3','cluster 4','cluster 5','cluster 6','cluster 7')).T
```

```python
completeness_score(classes[1],new_predicts)
```

```python
homogeneity_score(classes[1],new_predicts)
```

```python
#for compare
print "The completeness score before PCA: {}".format(completeness_score(classes[1],predicts))
print "The completeness score after PCA: {}".format(completeness_score(classes[1],new_predicts))

print '\n'

print "The homogeneity score before PCA: {}".format(homogeneity_score(classes[1],predicts))
print "The homogeneity score after PCA: {}".format(homogeneity_score(classes[1],new_predicts))
```
