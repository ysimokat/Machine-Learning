```python
## basic statistics for categorical data
df.groupby('pep').describe(include=object)
```

```python
## people who say 'yes' to PEP are older and have more income and have slightly fewer children
df.groupby('pep').mean()
```

```python
## male are more likely to say yes for buy PEP than female
df.groupby(['gender','pep'])['gender'].count()
```

```python
## the highest proportion of yes for buy PEP is in the suburbs and  
## the highest number of overall no for buy PEP is in the inner city
df.groupby(['region','pep'])['region'].count()
```

```python
## People who are not married turn to buy PEP
df.groupby(['married','pep'])['married'].count()
```

```python
## Owning a car does not appear to be a significant factor in whether people buy PEP or not
df.groupby(['car','pep'])['car'].count()
```

```python
## people who have a savings account are more likely to say no,
## but not having a savings account does not appear to be a significant factor in whether people buy PEP or not
df.groupby(['savings_acct','pep'])['savings_acct'].count()
```

```python
## Neither having or not having a checking account seems to favor buying PEP, 
## but people who do not have checking accounts are slightly more likely to buy PEP than those who do have checking accounts
df.groupby(['current_acct','pep'])['current_acct'].count()
```

```python
## Mortgage is not a significant factor in whether people buy PEP or not
df.groupby(['mortgage','pep'])['mortgage'].count()
```

```python
## z-score normalization to standardize the value of income attribute
income_std = (df['income'] - df['income'].mean()) / df['income'].std()
income_std.head(10)
```

```python
## make sure do not change the original income attribute in the table
df.head()
```

```python
## Discretize the age attribute into 3 categories
age_norm = pd.qcut(df.age, 3, labels=['young','mid-age','old'])
age_norm.head(10)
```

```python
## make sure do not change the original age attribute in the table
df.head(5)
```

```python
## use Min-Max Normalization to transform the values of all numeric attributes 
d = df[['age','income','children']]
df[['age','income','children']]=(d-d.min())/(d.max()-d.min())
```
